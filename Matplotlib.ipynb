```python
# Give a hike to people working more than 4 projects
```

```python
hr_data.columns
```

```python
def f(r):
    if r.number_project > 4:
        if r.salary == 'medium':
            return 'high'
        elif r.salary == 'low':
            return 'medium'
        else:
            return 'high'
    else:
        return r.salary
hr_data['new_salary'] = hr_data.apply(f,axis=1)
```

```python
new_salary = hr_data.new_salary.value_counts()
```

```python
old_salary = hr_data.salary.value_counts()
```

```python
salary_comp = pd.DataFrame({'Old Salary':old_salary, 'New Salary':new_salary})
```

```python
salary_comp.plot.bar()
```

## Assignment : Bucketize customers for targetting marketing

```python
import datetime
```

```python
#Assign Bronze
customer_churn_data[customer_churn_data.total_usage_days < datetime.timedelta(27,0,0)].head()
```

```python
#Assign Silver
customer_churn_data[(customer_churn_data.total_usage_days > datetime.timedelta(27,0,0)) & (customer_churn_data.total_usage_days < datetime.timedelta(110,0,0))].head()
```

```python
customer_churn_data['category'] = None
```

```python
def f(d):
    if d < datetime.timedelta(27,0,0):
        return 'Bronze'
    elif d > datetime.timedelta(27,0,0) and d < datetime.timedelta(110,0,0):
        return 'Silver'
    else:
        return 'Gold'
    
customer_churn_data['category']= customer_churn_data.total_usage_days.map(f)
```
